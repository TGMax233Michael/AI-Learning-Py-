### 折线图
### plot(x, y)
### 参数
marker          对每个点进行标记

markersize      标记的大小

linestyle       折线的样式

linewidth       折线的宽度

### 其他设置
title()     设置标题

x/ylabel()  x/y轴标签

x/ylim()    x/y轴范围

legend()    显示图例

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-np.pi, np.pi, 100)
y = np.sin(x)


plt.figure(figsize=(16, 10), dpi=80)
plt.plot(x, y, color="RED", marker="o", markersize=4, linestyle="-")
plt.title("Sine")
plt.xlabel("x")
plt.ylabel("y=sin(x)")
plt.legend()
plt.show()

```

### 散点图
### scatter(x, y)
### 参数
marker          对每个点进行标记

s               标记的大小


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-1, 1, 25)
y = np.arctan(x)


plt.figure(figsize=(16, 10), dpi=80)
plt.scatter(x, y, color="ORANGE", marker="*", s=200)
plt.title("Arctan")
plt.xlabel("x")
plt.ylabel("y=arctan(x)")
plt.legend()
plt.show()
```

### 柱状统计图
### hist(y)
### 参数
bins            分成多少个柱子

rwidth          柱子间的间距

range           显示的范围

orientation     水平/垂直(默认)

histtype        柱子的样式


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-1, 1, 1000)
y = np.arctan(x)


plt.figure(figsize=(16, 10), dpi=80)
plt.hist( y, color="LIGHTBLUE", bins=100, rwidth=1, range=(-1, 1))

plt.show()
```
